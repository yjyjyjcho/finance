```
# efficient frontier_효율적 투자선

import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
import seaborn  as sns
sns.set()
import scipy.optimize as sco

'''엑셀 파일 읽어온 후, 연평균 수익률로 환산하기'''
data =pd.read_excel('C:/Users/CHRIS/Desktop/Dong/Python/dataset.xlsx')
data = data.iloc[1:,:].astype(float) ; data.columns = ['Index', 'WTI','Gold']
data = np.log(data/data.shift(1)).shift(-1).dropna()


##우리 데이터##
# df_gaps1

'''Mu벡터, Vol벡터 구하기'''
n = len(data.columns) #자산군 갯수
Mu = DataFrame(np.empty((1,n)), columns = data.columns)

Mu['Index'] = data.iloc[:,0].mean()*12
for i in range(1,n) :
    Mu.iloc[:,i] = data.iloc[122:,i].mean()*12

Vol = DataFrame(np.empty_like(Mu), columns = data.columns)
for i in range(n) :
    Vol.iloc[:,i] = data.iloc[122:,i].std()*np.sqrt(12)


'''공분산 행렬 구하기'''
Cov = data.iloc[122:,:].cov()*12



def stat(weights):

    '''포트폴리오 통계치 출력

    인수
    ====
    weights : array-like
        포트폴리오 비중


    반환값
    ======

    Rp : float
        포트폴리오 수익률
    Rvol : float
        포트폴리오 변동성
    Rp/Rvol : float
        무위험 이자율이 0일때 샤프 지수
    '''
    weights = np.array(weights)
    Pr = np.dot(weights,Mu.T).sum()
    Pvol = np.sqrt(np.dot(np.dot(weights, Cov), weights.T))
    Sharp_ratio = (Pr/Pvol).sum()

    return np.array([Pr, Pvol, Sharp_ratio])



'''제약식(목표수익률)을 추가해서 각 수익률별 최저분산 포트폴리오의 집합을 구한다'''


#초기 비중 설정
W = np.random.random(n)
W = W/np.sum(W)

# 대상 함수 (분산)
def E_frontier(weight):
    return stat(weight)[1]**2

#범위 설정 (제약식[목표수익률]은 계속 변하므로 반복문 안에 정의한다.)
bnds = tuple((0,1) for x in range(n))


#최적화 
T_return = np.linspace(0.06,0.15,50) #수익률 구간 0~20%  
T_vol = []           #목표 수익률별 최저분산 포트폴리오의 표준편차가 들어갈 벡터

for r in T_return :
    cons = ({'type' : 'eq', 'fun' : lambda x : np.sum(x) - 1},   #np.sum(x) = 비중의 총합   
            {'type' : 'eq', 'fun' : lambda x : stat(x)[0] - r }) #Tr = 목표 수익률 
    opts = sco.minimize(E_frontier, W , method = 'SLSQP', bounds = bnds, constraints = cons) #제한 수익률에서 분산 minimize
    T_vol.append(np.sqrt(opts['fun']))

T_return = T_return        #목표 수익률 벡터
T_vol = np.array(T_vol)    #각각의 목표 수익률을 달성하는 최저분산 포트폴리오의 수익률 표준편차



'''효율적 투자선 그리기!!!'''


plt.figure(figsize=(12,6))

#효율적 투자선
plt.scatter(T_vol,T_return, c = T_return/T_vol, marker = 'X')


'''투자 가능한 모든 포트폴리오를 점으로 찍어보자'''

PR = []
PVOL = []

#1000개의 무작위 포트폴리오 만들기
for i in range(1000):
    weight = np.random.random(3)
    weight = weight/np.sum(weight)      #weight 

    PR.append(np.dot(weight,Mu.T).sum())
    PVOL.append(np.sqrt(np.dot(np.dot(weight,Cov),weight.T)))
    
PR = np.array(PR)
PVOL = np.array(PVOL)



#랜덤 포트폴리오
plt.scatter(PVOL,PR, c = PR/PVOL, cmap='viridis', marker = 'o' ,alpha = 0.3 )


'''최저분산 포트폴리오와 최대 샤프지수를 가진 포트폴리오도 찍어보자 '''

''' maximize sharp 포트폴리오'''

def f_maximize_sharp(w) :
    return -stat(w)[2] # -1 * 샤프지수 return

cons = ({'type' : 'eq' , 'fun' : lambda x : np.sum(x) - 1})
    
opts = sco.minimize(f_maximize_sharp,  W,  method = "SLSQP",
                    bounds = bnds, constraints = cons)

P_sharp_max = stat(opts['x'])   #Sharp지수를 maximize하는 포트폴리오 [return, vol, sharp]

#최대 샤프지수를 갖는 포트폴리오
plt.plot(P_sharp_max[1],P_sharp_max[0], marker='*', markersize= 25)


'''minimize 분산 포트폴리오'''

def f_min_var(w) :
    return stat(w)[1]**2

optv = sco.minimize(f_min_var, W, method = "SLSQP" , bounds = bnds, constraints = cons)


#최저 분산 포트폴리오
plt.plot(np.sqrt(optv['fun']),stat(optv['x'])[0], color = 'red', marker = "*", markersize = 25)




plt.xlabel('Expected Volatility')
Plt.ylabel('Expected Return')
plt.colobar(label = 'Sharp Ratio')
plt.show()

```
