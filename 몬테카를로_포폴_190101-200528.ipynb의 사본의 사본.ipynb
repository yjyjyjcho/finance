```
#####몬테카를로 시뮬레이션_포폴 수익, 리스크 최적화################################
# import quandl
import pandas as pandas
import numpy as np
import matplotlib.pyplot as plt


# ###########################################33
# sharpe ratio
# 단순 가정 위해 리스크 비율 0% 가정
# ##############################################

#get adjusted closing prices of 5 selected companies with Quandl
# quandl.ApiConfig.api_key = 'zcfj'
selected = ["ksp", "ksd", "snp", "uro", "nik", "csi", "bond", "comp", "hiy", "gold",
                   "oil", "inv", "dol", "dol_inv", "mon"]
# data = quandl.get_table('WIKI/PRICES', ticker = selected,
#                             qopts = {'columns': ['date', 'ticker', 'adj_close']},
#                             date = {'gte': '2014-1-', 'lte': '2016-12-31' }, paginate = True)

# #reorganize data pulled by setting date as index with
# #columns of tickers and their corresponding adjusted prices
# clean = data.set_index('date')
# table = clean.pivot(columns='ticker') #table : 내가 얻은 데이터프레임과 형태 동일

#caculate daily and annual returns of the stocks
returns_daily = df_gaps.pct_change()
returns_annual = returns_daily.mean() * 250

#get daily and covariance of returns of the stock
cov_daily = returns_daily.cov()
cov_annual = cov_daily * 250


#empty lists to strore returns, volatility and weights of imiginary portfolios
port_returns = []
port_volatility = []
stock_weights = []
sharpe_ratio =[]

# set the number of combinations for imaginary portfolios
num_assets = len(selected)
num_portfolios = 300000


#set random seed for reproduction's sake
np.random.seed(101)


#populate the empty lists with each portfolios returns, risk and weights
for single_portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    returns = np.dot(weights, returns_annual)
    volatility = np.sqrt(np.dot(weights.T, np.dot(cov_annual, weights)))

    sharpe = returns / volatility
    sharpe_ratio.append(sharpe)
  
    port_returns.append(returns)
    port_volatility.append(volatility)
    stock_weights.append(weights)

#a dictionary for Returns and Risk values of each portfolio
portfolio = {'Returns': port_returns,
                'Volatility': port_volatility,
                'Sharpe Ratio': sharpe_ratio}

# extend original dictionary to accomodate each ticker and weight in the portfolio
for counter,symbol in enumerate(selected):
    portfolio[symbol+' weight'] = [weight[counter] for weight in stock_weights]

#make a nice dataframe of the extended dictionary
df = pd.DataFrame(portfolio)

print(df)

#get better labels for desired arrangement of columns
columns_order = ['Returns', 'Volatility', 'Sharpe Ratio'] + [stock+' weight' for stock in selected]

#reorder dataframe columns
df = df[columns_order]

#######우리 비중에 맞는 인덱스 찾기######

def mse(y,t):
  return np.sum((y-t)**2)

df_inputratio = pd.DataFrame({'ksp weight' : [0], 'ksd weight' : [0],
      'snp weight' : [0], 'uro weight' : [0], 'nik weight' : [0], 'csi weight' : [0],
        'bond weight' : [0],'comp weight' : [0], 'hiy weight' : [0], 'gold weight' : [0], 
        'oil weight' : [0], 'inv weight' : [0],'dol weight' : [0], 'dol_inv weight' : [0],
        'mon weight' : [0]}, index=[0])
    
for i in range(df_inputratio.shape[0]):
  for j in range(df_inputratio.shape[1]):
        df_inputratio.iloc[i, j] = 0

def we_ratio(kspweight, ksdweight,
        snpweight, uroweight, nikweight, csiweight, bondweight,
        compweight, hiyweight, goldweight, oilweight, invweight,
        dolweight, dol_invweight, monweight):
    # print("비율을 입력해주세요" '\n')
    # kspweight = input("ksp weight")
    # ksdweight = input("ksd weight")
    # snpweight = input("snp weight")
    # uroweight = input("uro weight")
    # nikweight = input("nik weight")
    # csiweight = input("csi weight")
    # bondweight = input("bond weight")
    # compweight = input("comp weight")
    # hiyweight = input("hiy weight")
    # goldweight = input("gold weight")
    # oilweight = input("oil weight")
    # invweight = input("inv weigh")
    # dolweight = input("dol weight")
    # dol_invweight = input("dol_inv weight")
    # monweight = input("mon weight")




    min1 = 1
    mse_sum = 0
    for i in range(df.shape[0]):
      for j in range(3, 18):
        y = df.iloc[i,j]
        t = df_inputratio.iloc[0, j-3]
        mse_sum += mse(y, t)
      if min1 > mse_sum:
        min1 = mse_sum
        row = i

    return row, df.iloc[row,0], df.iloc[row,1], df.iloc[row,2]

# result = we_ratio(0.15, 0.10, 0.20, 0, 0, 0.15, 0, 0, 0.34, 0.05, 0, 0, 0, 0, 0.01)
# print(result)
print('\n')
result2 = we_ratio(0.2, 0.05, 0.15, 0, 0, 0.2, 0, 0, 0.24, 0.15, 0, 0, 0, 0, 0.01)
print(result2)

####plot the efficient frontier with a scatter plot
plt.style.use('seaborn-dark')
df.plot.scatter(x='Volatility', y='Returns', c='Sharpe Ratio',
                cmap = 'RdYlGn', edgecolors='black', figsize=(10, 8), grid=True)
plt.xlabel('Volatility (Std. Deviation)')
plt.ylabel('Expected Returns')
plt.title('Efficient Frontier')
plt.show()



####optimal answer####
#find min volatility & max sharpe values in the dataframe(df)
min_volatility = df['Volatility'].min()
max_sharpe = df['Sharpe Ratio'].max()

#use the min, max values to locate and create the two special portfoilos
sharpe_portfolio = df.loc[df['Sharpe Ratio'] == max_sharpe]
min_variance_port = df.loc[df['Volatility'] == min_volatility]

# 우리 포폴 찾기
we_port = df.iloc[0]

#plot frontier, max sharpe & min Volatility values with a scatterplot
plt.style.use('seaborn-dark')
df.plot.scatter(x='Volatility', y='Returns', c='Sharpe Ratio', 
                cmap = 'RdYlGn', edgecolors='black', figsize = (10, 8), grid = True)
plt.scatter(x=sharpe_portfolio['Volatility'], y=sharpe_portfolio['Returns'], c='red', marker='D')
plt.scatter(x=min_variance_port['Volatility'], y=min_variance_port['Returns'], c='blue', marker='D')
plt.scatter(x=we_port['Volatility'], y=we_port['Returns'], c='pink', marker='D') ##우리꺼
plt.xlabel('Volatility (Std. Deviation)')
plt.ylabel('Expected Returns')
plt.title('Efficient Frontier')
plt.show()
```
